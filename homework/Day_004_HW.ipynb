## 將下列部分資料片段 sub_train 使用 One Hot encoding, 並觀察轉換前後的欄位數量 (使用 shape) 與欄位名稱 (使用 head) 變化

```python
all_combination = list(set(sub_train['WEEKDAY_APPR_PROCESS_START']))

monday = [1 if item == 'MONDAY' else 0 for item in sub_train['WEEKDAY_APPR_PROCESS_START'].tolist()]
tuesday = [1 if item == 'TUESDAY' else 0 for item in sub_train['WEEKDAY_APPR_PROCESS_START'].tolist()]
wednesday = [1 if item == 'WEDNESDAY' else 0 for item in sub_train['WEEKDAY_APPR_PROCESS_START'].tolist()]
thursday = [1 if item == 'THURSDAY' else 0 for item in sub_train['WEEKDAY_APPR_PROCESS_START'].tolist()]
friday = [1 if item == 'FRIDAY' else 0 for item in sub_train['WEEKDAY_APPR_PROCESS_START'].tolist()]
satursday = [1 if item == 'SATURDAY' else 0 for item in sub_train['WEEKDAY_APPR_PROCESS_START'].tolist()]
sunday = [1 if item == 'SUNDAY' else 0 for item in sub_train['WEEKDAY_APPR_PROCESS_START'].tolist()]
```

```python
hot_df = pd.DataFrame(dict(zip(all_combination, [sunday, wednesday, satursday, friday, tuesday, monday, thursday])))
hot_df
```
